```python

#def limites(base_airbnb_2['price']):
def limites(coluna):

    q1 = coluna.quantile(0.25)
    q3 = coluna.quantile(0.75)
    amplitude = q3 -q1
    limite_inferior= q1 -1.5*amplitude
    limite_superior= q3 +1.5*amplitude
    return limite_inferior, limite_superior


#q1 25% e q3 75%
# amplitude q1-q3
#limite inferior q1-1,5*amplitude
#limite superior q3+1,5*amplitude
```

```python
#saber o limite inferior e o superior
def excluir_outliers(df, nome_coluna):
      qtde_linhas = df.shape[0]
      lim_inf, lim_sup = limites(df[nome_coluna])
      #df= df.loc[linha, coluna]
      #df= df.loc[linhas, :]
      df= df.loc[(df[nome_coluna] >= lim_inf) & (df[nome_coluna] <= lim_sup), :]
      linhas_removidas =  qtde_linhas  - df.shape[0]
      return df, linhas_removidas
```

## Função Modelo de Previsão


comparar os modelos em duas métricas

- Métricas de avaliação  

```python

def avaliar_modelo(nome_modelo, y_teste, previsao):
    #eixo x =  valores
    #eixo y = preço reais dos imóveis que eu quero prever  
    #previsão = modelo de previsão do preço 
    
    #correlação R2
    r2 = r2_score(y_teste, previsao)
    #erro quadrático médio gerado pela biblioteca do machine learning 
    # raiz do erro quadrático médio que iremos usar RSME
    RMSE = np.sqrt(mean_squared_error(y_teste, previsao))
    
    return f'Modelo {nome_modelo}:\nR2:{r2}\nRSME:{RMSE}'
    
    
    
```

# Análise do Melhor Modelo

```python
def avaliar_modelo_porcentagem(nome_modelo, y_teste, previsao):
    #eixo x =  valores
    #eixo y = preço reais dos imóveis que eu quero prever  
    #previsão = modelo de previsão do preço 
    
    #correlação R2
    r2 = r2_score(y_teste, previsao)
    #erro quadrático médio gerado pela biblioteca do machine learning 
    # raiz do erro quadrático médio que iremos usar RSME
    RSME = np.sqrt(mean_squared_error(y_teste, previsao))
    
    return f'Modelo {nome_modelo}:\nR2:{r2:.2%}\nRSME:{RSME:.2f}'
    
```

```python
for nome_modelo, modelo in modelos.items():
    print(nome_modelo,modelo)
    #treinar
    #modelo.fit(X_train, y_train)
    #testar => função : avaliar_modelo()
    #resposta => string
    # definir a previsão
    previsao = modelo.predict(X_test)
    print(avaliar_modelo_porcentagem(nome_modelo, y_test, previsao))
```

#### Modelo de ExtraTrees

Maior valor de R2 (valor de acerto)

e o menor valor de RSME (errar menos)


- Modelo Escolhido ocmo Melhor Modelo:  ExtraTressRegressor

 Esse foi o modelo com maior valor de R² e ao mesmo tempo o menor valor de RSME. 
 
 Como não tivemos uma grande diferença de velocidade de treioon e de previsão desse modelo com o modelo de RandomForest (que teve resultados próximos de R² e RSME, vamos escolher o Modelo de ExtraTrees. 
 
  O Modelo de Regressão linear não obteve um resultado satisfatório, com valores de R² e RSME muito piores do que o modelo ExtraTrees, por exemplo. 
  
  R² seria o melhor para tomada de decisão, mas usamos a segunda variável, o erro quadrático médio
