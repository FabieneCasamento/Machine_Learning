# Machine Learning project

```python
X= df_music.drop(columns=['genre'])
display(X)
```

```python
Y = df_music['genre']
display(Y)
```

```python
from sklearn.tree import DecisionTreeClassifier
#modelo para treinar o padrão
model =DecisionTreeClassifier()
model.fit(X,Y)
```

```python
df_music
```

```python
#modelo preditor
predictions = model.predict([[21,1], [22,0]])
```

```python

display(predictions)
```

```python
print('previsão para quem tem 21 anos e é homem (1) e 22 anos e é mulher (0)')
print()
print('tem 21 anos e é homem (1), que tipo de música escuta:')
print(predictions[0])
print()
print('22 anos e é mulher (0), que tipo de música escuta:')
print(predictions[1])
```

```python
from sklearn.model_selection import train_test_split

#treinar e testar
#70 a 80% dos dados treinamos e os outros 30% testamos 

#20% dos dados testamos neste caso
X_train, X_test, y_train, y_test  =train_test_split(X,Y,test_size=0.2)


from sklearn.metrics import accuracy_score
#modelo preditor
#modelo para treinar o padrão
#X sem genre
#Y só com o genre (só com o gênero da música)
model =DecisionTreeClassifier()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
display(predictions)

score = accuracy_score(y_test, predictions)
display(score)
print(X_test)
```

```python
df.isnull().sum()
```

```python
dados_sem_nulos = df.dropna(subset=['Year'])
dados_sem_nulos
```

```python
dados_sem_nulos.isnull().sum()
#soma de dados nulos no dataframe
```

```python
dados_sem_nulos = dados_sem_nulos.dropna(subset=['Publisher'])
dados_sem_nulos
```

```python
dados_sem_nulos= dados_sem_nulos.sort_values(by='Year', ascending= True)
display(dados_sem_nulos)
```

```python
df_vendas_X= dados_sem_nulos[['Year','Global_Sales']]
display(df_vendas_X)

df_vendas_Y= dados_sem_nulos[['Publisher']]
display(df_vendas_Y)
```

```python
print(df['Publisher'].unique())
print(df['Year'].unique())
```

```python
df_vendas_X.info()
```

```python
from sklearn.model_selection import train_test_split

#treinar e testar
#70 a 80% dos dados treinamos e os outros 30% testamos 

#20% dos dados testamos neste caso
X_train_j, X_test_j, y_train_j, y_test_j  =train_test_split(df_vendas_X,df_vendas_Y,test_size=0.2)


from sklearn.metrics import accuracy_score
#modelo preditor
#modelo para treinar o padrão
#X ano e vendas globais
#Y só com a marca do jogo
model =DecisionTreeClassifier()
model.fit(X_train_j,y_train_j)
predictions = model.predict(X_test_j)
display(predictions)

score = accuracy_score(y_test_j, predictions)
display(score)
print(X_test_j)
```
