#### Com isso, podemos partir para a preparação dos dados para treinarmos o Modelo de Machine Learning

- Separando em dados de treino e dados de teste

```python
#sklearn
#separar, treino e teste
from sklearn.model_selection import train_test_split

#x = valores que eu quero usar, tv, radio e jornal, todas as informações, mesmo vendas (drop jogar fora vendas, pego tudo)
#y = valores vendas


x = tabela.drop("Vendas", axis =1)
y = tabela["Vendas"]

#0.3 30% para teste, ai o treino é 70%

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python

# problema de regressão
#regressão linear, número
#random forest, categoria (árvore de decisão), classificar


from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

#criei uma e outra, criar a inteligência artificial

linear = LinearRegression() 
randomforest = RandomForestRegressor()

#treinar a ia
linear.fit(x_treino, y_treino)
randomforest.fit(x_treino, y_treino)

#árvore de decisão
#cálculo dados


#run treinou
```

```python
#---------------------
#teste da al e avaliação do melhor modelo

#r2, 0 a 100%, mais em 100%, melhor e mse(erro quadrático médio)
from sklearn import metrics

#prevsão nos dados no teste, sem testado
teste_linear = linear.predict(x_teste)
teste_random = randomforest.predict(x_teste)

#r2
r2_linear = metrics.r2_score(y_teste, teste_linear)

#valores corretos, previsão


r2_random = metrics.r2_score(y_teste, teste_random)

print(r2_linear, r2_random)

#erro médio

erro_linear = metrics.mean_squared_error(y_teste, teste_linear)
erro_random = metrics.mean_squared_error(y_teste, teste_random)

print(erro_linear, erro_random)

#random é melhor
#errou menos
```

#### depois da preparação dos dados para treinarmos o Modelo de Machine Learning

- Separando em dados de treino e dados de teste

#### Temos um problema de regressão - Vamos escolher os modelos que vamos usar:

- Regressão Linear
- RandomForest (Árvore de Decisão)

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
import numpy as np

# treino AI
lin_reg = LinearRegression()
lin_reg.fit(x_treino, y_treino)

rf_reg = RandomForestRegressor()
rf_reg.fit(x_treino, y_treino)
```

#### Teste da AI e Avaliação do Melhor Modelo

- Vamos usar o R² -> diz o % que o nosso modelo consegue explicar o que acontece
- Também vamos olhar o MSE (Erro Quadrático Médio) -> diz o quanto o nosso modelo "erra" quando tenta fazer uma previsão

```python
# teste AI
test_pred_lin = lin_reg.predict(x_teste)
test_pred_rf = rf_reg.predict(x_teste)

r2_lin = metrics.r2_score(y_teste, test_pred_lin)
mse_lin = metrics.mean_squared_error(y_teste, test_pred_lin)
print(f"R² da Regressão Linear: {r2_lin}")
print(f"RMSE da Regressão Linear: {mse_lin}")
r2_rf= metrics.r2_score(y_teste, test_pred_rf)
mse_rf = metrics.mean_squared_error(y_teste, test_pred_rf)
print(f"R² do Random Forest: {r2_rf}")
print(f"RMSE do Random Forest: {mse_rf}")
```

```python
rmse_lin = np.sqrt(metrics.mean_squared_error(y_teste, test_pred_lin))
print(f"*RMSE da Regressão Linear: {rmse_lin}")
rmse_rf = np.sqrt(metrics.mean_squared_error(y_teste, test_pred_rf))
print(f"*RMSE do Random Forest: {rmse_rf}")
print()
r2_lin = metrics.r2_score(y_teste, test_pred_lin)
mse_lin = metrics.mean_squared_error(y_teste, test_pred_lin)
print(f"R² da Regressão Linear: {r2_lin}")
print(f"RMSE da Regressão Linear: {mse_lin}")
r2_rf= metrics.r2_score(y_teste, test_pred_rf)
mse_rf = metrics.mean_squared_error(y_teste, test_pred_rf)
print(f"R² do Random Forest: {r2_rf}")
print(f"RMSE do Random Forest: {mse_rf}")
```
